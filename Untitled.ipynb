# Step 1: Data Wrangling and Cleansing
## Load CSV string into DataFrame

```python
df = pd.read_csv("Uberf Data/UberDataset.csv")
```

```python
# Display initial data info
print("Initial Data Info:")
print(df.info())
```

```python
print("\nInitial Missing Values:")
print(df.isnull().sum())
```

```python
# Clean data
# Convert date columns to datetime, handling mixed formats
df['START_DATE'] = pd.to_datetime(df['START_DATE'], format='mixed', errors='coerce')
df['END_DATE'] = pd.to_datetime(df['END_DATE'], format='mixed', errors='coerce')
```

```python
# Replace empty or NaN PURPOSE with 'Unknown'
df['PURPOSE'] = df['PURPOSE'].fillna('Unknown')
```

```python
df
```

## Verify no missing values

```python

print("\nCleaned Data Info:")
print(df.info())
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
```

# Step 2: Data Transformation (ETL)

```python

# Calculate trip duration in minutes
df['DURATION_MIN'] = (df['END_DATE'] - df['START_DATE']).dt.total_seconds() / 60
```

```python
df.drop('DURATION', axis=1, inplace=True)
```

```python
# Extract day of week and hour
df['DAY_OF_WEEK'] = df['START_DATE'].dt.day_name()
df['HOUR'] = df['START_DATE'].dt.hour
```

```python
# Categorize trips by miles
df['TRIP_TYPE'] = pd.cut(df['MILES'], bins=[0, 5, 20, float('inf')], 
                         labels=['Short', 'Medium', 'Long'])
```

```python
# Create transformed DataFrame
transformed_df = df.copy()
```

# Display transformed data head

```python

print("\nTransformed Data Head:")
print(transformed_df.head())
```

# Statistical summary

```python

print("\nStatistical Summary:")
print(transformed_df.describe())
```

# Step 3: Exploratory Analysis

```python

# Summary by CATEGORY
category_summary = transformed_df.groupby('CATEGORY').agg({
    'MILES': ['sum', 'mean', 'count'],
    'DURATION_MIN': ['sum', 'mean']
}).round(2)
print("\nSummary by Category:")
print(category_summary)
```

# Summary by PURPOSE

```python

purpose_summary = transformed_df.groupby('PURPOSE').agg({
    'MILES': ['sum', 'mean', 'count'],
    'DURATION_MIN': ['sum', 'mean']
}).round(2)
print("\nSummary by Purpose:")
print(purpose_summary)
```

# Key metrics

```python

total_trips = len(transformed_df)
total_miles = transformed_df['MILES'].sum()
avg_miles = transformed_df['MILES'].mean()
business_pct = (transformed_df['CATEGORY'] == 'Business').mean() * 100
print(f"\nKey Metrics:")
print(f"Total Trips: {total_trips}")
print(f"Total Miles: {total_miles:.1f}")
print(f"Average Miles: {avg_miles:.2f}")
print(f"Business Trips Percentage: {business_pct:.2f}%")
```

# Step 4: Data Visualization


```python
# 2. Pie Chart of Categories
plt.figure(figsize=(8, 8))
transformed_df['CATEGORY'].value_counts().plot.pie(
    autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'], startangle=90
)
plt.title('Trip Categories')
plt.ylabel('')
plt.show()
```

```python
# 3. Bar Plot of Purposes
plt.figure(figsize=(12, 6))
sns.countplot(y='PURPOSE', data=transformed_df, 
              order=transformed_df['PURPOSE'].value_counts().index, 
              palette='viridis')
plt.title('Number of Trips by Purpose')
plt.xlabel('Count')
plt.ylabel('Purpose')
plt.show()
```

```python
# 4. Box Plot of Miles by Day of Week
plt.figure(figsize=(10, 6))
sns.boxplot(x='DAY_OF_WEEK', y='MILES', data=transformed_df, palette='Set3')
plt.title('Miles by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Miles')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Bar Plot of Average Miles and Duration by Purpose
summary = transformed_df.groupby('PURPOSE').agg({'MILES': 'mean', 'DURATION_MIN': 'mean'}).reset_index()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='MILES', y='PURPOSE', data=summary, palette='viridis')
plt.title('Average Miles by Purpose')
plt.xlabel('Average Miles')
plt.ylabel('Purpose')
```

```python
plt.subplot(1, 2, 2)
sns.barplot(x='DURATION_MIN', y='PURPOSE', data=summary, palette='magma')
plt.title('Average Duration by Purpose')
plt.xlabel('Average Duration (minutes)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# 6. Trip Type Distribution
plt.figure(figsize=(8, 6))
transformed_df['TRIP_TYPE'].value_counts().plot.bar(color='teal')
plt.title('Trip Type Distribution')
plt.xlabel('Trip Type')
plt.ylabel('Count')
plt.show()
```
